# List comprehensions

```python
c = [<var expr.> for <var> in <iterable/iterator> if <boolean expression using var>]
```

````python
c = [num for num in range(10001) if num % 2 == 0]
print(c)
````

````python
squares_of_odds = [i**2 for i in range(1000) if i % 2 == 1]
squares_of_odds
````

````python
import math
math.sqrt(998001)
````

````python
cartesian_coords = [(x, y) for x in range(100) for y in range(100)]
print(cartesian_coords)
````

````python
cartesian_coords = []

for x in range(100):
    for y in range(100):
        cartesian_coords.append((x, y))
cartesian_coords
````

````python
table = []

for row in range(3):
    row_list = []
    for i in range(row*3, row*3 + 3):
        row_list.append(i)
    table.append(row_list)
table
````

````python
table = []
size = 22

for row in range(size):
    row_list = []
    for i in range(row*size, row*size + size):
        row_list.append(i)
    table.append(row_list)

for row in table:
    print(row)
````

````python
for row in table:
    for i, col in enumerate(row):
        row[i] += i

for row in table:
    print(row)
````

````python
for i in range(len(table)):
    for j in range(len(table[i])):
        table[i][j] *= 5

for row in table:
    print(row)
````
