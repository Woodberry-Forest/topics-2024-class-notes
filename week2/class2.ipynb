# Dictionaries

```python
grades = {"Corey": 94, "Eric": 29, "Abraham": 1, "Cody": 99}
```

```python
for key in grades.keys():
    print(key, grades[key])
```

```python
grades.values()
```

```python
for key, value in grades.items():
    print(key, ": ", value, sep="")
```

```python
grades["Liam"] = 86
```

```python
grades["Penn"] = 54
grades["Ben"] = 43
grades
```

```python
with open("./gatsby.txt", "r", encoding="utf-8") as f:
    text = f.read()

print(text)
```

```python
len(text)
```

```python
letters = {}

for letter in text:
    if letter not in letters.keys():
        letters[letter] = 1
    else:
        letters[letter] += 1

letters
```

```python
ll = list(letters.items())
```

```python
ll.sort(key=lambda entry: entry[1], reverse=True)
ll
```
